```python
import glob
def preprocessing(input_path,start):
  all_input_list = sorted(glob.glob(input_path))
  train = pd.DataFrame()
  casee = start
  for datapath in all_input_list:
    casee += 1
    data = pd.read_csv(datapath) 
    data['obs_time'] = data.index % 24 
    data = abs(data)
    col_list = data.columns
    for i in range(0,len(col_list)):
      col = col_list[i]    
      if '누적' in col :
        data[col] = data.groupby((data.obs_time == 0).cumsum()).agg(col_list[i-1]).cumsum()
    data['Case'] = casee
    train = pd.concat([train,data])
  return train
```

```python
input_path1 = '/content/drive/MyDrive/dacon/gan/상추/train_input/*.csv'
input_path2 = '/content/drive/MyDrive/dacon/gan/상추/test_input/*.csv'
```

```python
train = preprocessing(input_path1,0)
test = preprocessing(input_path2,30)
```
